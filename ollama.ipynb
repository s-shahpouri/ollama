# Making API Requests to Ollama Server
### Remote access, Scalable, Best for web-based systems

```python
import requests
import json

# Set the URL of the API
url = "http://127.0.0.1:11434/api/chat" 

# Define the payload for the POST request
payload = {
    "model": "deepseek-r1:8b",
    "messages": [{"role": "user", "content": "What is Python?"}]
}

# Send the POST request with the payload
response = requests.post(url, json=payload, stream=True)
print(requests)
# Check if the request was successful
if response.status_code == 200:
    print("Response received successfully")

    # Iterate over the lines in the response stream
    for line in response.iter_lines(decode_unicode=True):
        if line:
            try:
                # Parse the JSON data from the response line
                json_data = json.loads(line)
                
                # Check if the message key exists and print the content
                if "message" in json_data and "content" in json_data["message"]:
                    print(json_data["message"]["content"], end="")

            except json.JSONDecodeError:
                # Handle any JSON parsing errors
                print("Error decoding JSON")

else:
    # If the request was not successful, print the status code
    print(f"Request failed with status code {response.status_code}")
```

# Using Ollama Python Client

### Direct Interaction, Quick Setup, Local Dependency, Late Respond.

```python
import ollama

Client = ollama.Client()


model = "deepseek-r1:8b"
prompt = "What is python?"

response = Client.generate(model = model, prompt=prompt)

print(response.response)
```

Usage: Run the model with ollama run SERA to interact and assist with medical inquiries.

```python

# Initialize the Ollama client
client = ollama.Client()

# Define the model name (in this case, your custom model 'SERA')
model_name = "SERA"

# Define the prompt you want to ask the model
prompt = "How can I get an appointment from the hospital?"

# Generate a response from the model
response = client.generate(model=model_name, prompt=prompt)
print(response.response)
```
